#### Can we establish some kind of relationship between pandemic outbreaks and the performance of indices like S&P500?
We intend throughout this notebook to observe 4 pandemics that reached a substantial number of deaths.

## Asian Flu (1957)


The flu killed approximately 1 million to 2 million people worldwide. 


```python
# data = [
#         {'startDate': datetime.strptime("1957-01-01", '%Y-%m-%d'), 
#          'announce': 'Asian Flu'},
#             ]  
# announcesdf = pd.DataFrame(data) 

# raw_data = te.fetchMarkets(symbol='SPX:IND', initDate='1955-01-01', endDate='1958-12-31', output_type='df')
# treated_data = raw_data.drop(['symbol','high','low','open'], axis=1)
# treated_data=treated_data.rename(columns={'close':'price'})
# fig, ax = plt.subplots(figsize=(20, 10))
# treated_data.plot(ax=ax)

# # Add labels to the plot
# style = dict(size=15, color='black')
# ax.text('1957-1-1', 48, "Asian Flu", **style)

raw_data = te.fetchMarkets(symbol='SPX:IND', initDate='1955-01-01', endDate='1958-12-31', output_type='df')
treated_data = raw_data.drop(['symbol','high','low','open'], axis=1)
treated_data=treated_data.rename(columns={'close':'price'})

def plot(df):
    data = [
        {'startDate': datetime.strptime("1957-01-01", '%Y-%m-%d'), 
         'announce': 'Asian Flu'},
            ]  
    announcesdf = pd.DataFrame(data) 
    fig = go.Figure()
    labels = df.index.tolist()
    line1 = df['price'].values.tolist()
    fig, ax1 = plt.subplots(1,1,figsize=(14,8), dpi= 80)
    for index, row in announcesdf.iterrows():
        #print(row['c1'], row['c2'])
        plt.plot([row['startDate'],row['startDate']],[0,55], lw=3, color ='khaki',zorder=-2)
    ax1.plot(labels, line1, color='darkred')
    ax1.set_xlabel('Date', fontsize=12)
    ax1.tick_params(axis='x', rotation=0, labelsize=12)
    ax1.set_ylabel('Price', color='darkred', fontsize=14)
    ax1.tick_params(axis='y', rotation=0, labelcolor='darkred' )
    ax1.grid(alpha=.3)
    ax1.set_title("S%P 500 and Asian Flu", fontsize=16)
    #ax1.legend()
    fig.tight_layout()
plot(treated_data)
```

## SARS (2002)
While SARS broke out in Nov. 2002, it did not begin to affect markets until March 2003 after the Chinese authorities reported the SARS outbreak to the World Health Organization. (https://investors-corner.bnpparibas-am.com/investing/coronavirus-lessons-from-sars-for-investors/)

## EBOLA (2013)
The Ebola epidemic in West Africa caused 11,310 deaths in Guinea, Liberia, and Sierra Leone, and 15 deaths outside of these three countries. Between December 2013 and February 2014 the S&P 500 fell 5.9%. The sectors most hurt by the epidemic were airlines, cruises, and hotels.

 Shares of American Airlines and Delta Air Lines both dropped 20% in October 2014 after news that an Ebola patient flew the day before being diagnosed.
 
 https://www.cdc.gov/vhf/ebola/history/2014-2016-outbreak/index.html
 
 https://money.cnn.com/2014/10/15/investing/ebola-spooks-wall-street-investors-stocks/index.html

## Conclusion

It seems to be difficult to determine to what degree these outbreaks interfered with the market. Over the last few decades, we have had a variation in the way technology and science forward the protocols for containing outbreaks, and we also have the growth and transformation from an economy restricted to certain parts of the world to a globalized economic model. It seems possible to suggest that SARS and COVID rather than EBOLA had a large spread area and ended up generating a period of economic downturn. And it also seems to be possible to say that such moments of outbreaks took place at times when there was also a strong economic downturn accompanied by a recovery in less than 10 years.
