# Stock Market Top & Bottom 5 perfomers

```python
import tradingeconomics as te
from num2words import num2words
import matplotlib.pyplot as plt
import pandas as pd
```

```python
te.login()
```

```python
te.getMarketsSearch(country='usa', output_type='df')
```

```python
# Dow Jones - INDU:IND
# S&P 500 - US500
index = te.getMarketsComponents(symbols='SPX:IND', output_type='df')
index
```

```python
sorted_index = index.sort_values(by='YTDPercentualChange')
sorted_index
```

#### Bottom 5 perfomers

```python
five_worst_perfomers = sorted_index.head()
five_worst_perfomers
```

```python
value_lost = 0

df = pd.DataFrame(columns=['Name', 'YTDPercentualChange'])
for index, row in five_worst_perfomers.iterrows():
    
    current_mark_cap = row['MarketCap']
    current_price = row['Close']
    num_outstanding_shares = current_mark_cap / current_price
    start_year_mark_cap = row['startYear'] * num_outstanding_shares    
    
    val = current_mark_cap - start_year_mark_cap

    df = pd.concat([df, pd.DataFrame({'Name': [row['Name']], 'YTDPercentualChange': [row['YTDPercentualChange']]})])
 
    value_lost += val
    
value_lost
df.reset_index(drop=True)
```

```python
value_lost
```

```python
'Value Lost: ' + num2words(abs(value_lost))
```

```python
five_best_perfomers = sorted_index.tail()
five_best_perfomers
```

```python
value_gained = 0

df_2 = pd.DataFrame(columns=['Name', 'YTDPercentualChange'])
for index, row in five_best_perfomers.iterrows():
    
    current_mark_cap = row['MarketCap']
    current_price = row['Close']
    num_outstanding_shares = current_mark_cap / current_price
    start_year_mark_cap = row['startYear'] * num_outstanding_shares    
    
    val = current_mark_cap - start_year_mark_cap

    df_2 = pd.concat([df_2, pd.DataFrame({'Name': [row['Name']], 'YTDPercentualChange': [row['YTDPercentualChange']]})])

    value_gained += val
    
value_gained
df_2.reset_index(drop=True)
```

```python
value_gained
```

```python
'Value Gained: ' + num2words(value_gained)
```

```python
fig = plt.figure(figsize=(20, 10))
ax = plt.subplot(111)
p1 = ax.bar(df['Name'], df['YTDPercentualChange'], width=0.8, color='r')
p2 = ax.bar(df_2['Name'], df_2['YTDPercentualChange'], width=0.8, color='g')

ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

ax.axhline(0, color='black', linewidth=1)
ax.set_ylabel('Year to date % change', fontsize=20)
ax.set_xlabel('Company Name', fontsize=20)

ax.set_title('Year to Date % Change - Top & Bottom 5', fontsize=22);

plt.setp(ax.get_xticklabels(), rotation=90, ha='right')

ax.bar_label(p1, padding=5, fontsize=14);
ax.bar_label(p2, padding=5, fontsize=14);
```
