Importing dependencies

```python
import tradingeconomics as te
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

API login

```python
te.login(); # Signup for a free API key at https://developer.tradingeconomics.com/
```

Importing data

```python
data = te.getWBHistorical(series_code = 'AGO.1.1_ACCESS.ELECTRICITY.TOT', output_type='df')
```

Plotting

```python
dates = pd.DatetimeIndex(data['date'])
labels = dates.year
values = data['value']

fig1, ax1 = plt.subplots(figsize=(10, 10))

ax1.pie(values, autopct='%1.1f%%',
        startangle=160)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

ax1.legend(labels,
          title="Year",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))


ax1.set_title("Angola's access to electricity (% of total population)")

plt.show()
```
