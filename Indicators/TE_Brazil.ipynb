# Brazil's correlation between Ecomic Growth and Social Development

## Brazil GDP Composition and Historical Performance

```python
indicators_list = ['GDP From Agriculture',
                  'GDP From Construction',
                  'GDP From Manufacturing',
                  'GDP From Mining',
                  'GDP From Public Administration',
                  'GDP From Services',
                  'GDP From Transport',
                  'GDP From Utilities',]

df_1996 = te.getHistoricalData(country=['Brazil'], indicator=indicators_list, initDate='1996-07-01', endDate='1996-12-30', output_type='df')
df_2000 = te.getHistoricalData(country=['Brazil'], indicator=indicators_list, initDate='2000-07-01', endDate='2000-12-30', output_type='df')
df_2010 = te.getHistoricalData(country=['Brazil'], indicator=indicators_list, initDate='2010-07-01', endDate='2010-12-30', output_type='df')
df_2015 = te.getHistoricalData(country=['Brazil'], indicator=indicators_list, initDate='2015-07-01', endDate='2015-12-30', output_type='df')
df_actual = te.getHistoricalData(country=['Brazil'], indicator=indicators_list, initDate='2021-05-01', endDate='2021-07-30', output_type='df')


fig = make_subplots(3, 2, specs=[[{'type':'domain'}, {'type':'domain'}],[{'type':'domain'}, {'type':'domain'}],[{'type':'domain'}, {'type':'domain'}]],
                    subplot_titles=['1996','2000', '2010','2015','Actual'])
fig.add_trace(go.Pie(labels=df_1996['Category'], values=df_1996['Value'], scalegroup='one',
                     name="2000"), 1, 1)
fig.add_trace(go.Pie(labels=df_2000['Category'], values=df_2000['Value'], scalegroup='one',
                     name="2000"), 1, 2)
fig.add_trace(go.Pie(labels=df_2010['Category'], values=df_2010['Value'], scalegroup='one',
                     name="2010"), 2, 1)
fig.add_trace(go.Pie(labels=df_2015['Category'], values=df_2015['Value'], scalegroup='one',
                     name="2015"), 2, 2)
fig.add_trace(go.Pie(labels=df_actual['Category'], values=df_actual['Value'], scalegroup='one',
                     name="Actual"), 3, 1)
fig.update_layout(height=800, width=1000,title_text="Brazil's GDP Composition - GPD composition along the time do not shows great changes")
fig.show(renderer='plotly_mimetype')

df_brazil_gdp = te.getHistoricalData(country='brazil', indicator='gdp', initDate='1990-01-01', output_type='df')
df_brazil_gdp = df_brazil_gdp[['DateTime','Value']] 
df_brazil_gdp=df_brazil_gdp.rename(columns={'Value':'USD Billion'})
fig = px.line(df_brazil_gdp, x='DateTime', y='USD Billion', title="Brazil's GDP Historical Performance")

fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)
fig.show(renderer='plotly_mimetype')
```

## If, on the one hand, we can say that the composition of Brazilian GDP has changed little over the last 30 years, the same cannot be said of the total value of GDP. It reached the highest value in December 2011, namely, 2,616.2 USD Billion, and in December 2019, before COVID-19, the value was decreasing to 1,877.95 USD Billion.

## On the other hand, looking at the indicators that can help us to verify to what extent the citizens of this country may be receiving direct and indirect benefits from the economic scenario that we see in the other indicators, what calls our attention is that social security is stagnant , the average monthly salary of Brazilian citizens, against the economic scenario had an increase between 2014 and 2015, but already in 2016 it was low. And it goes up again between July 2020, and thereafter in a downward movement. It seems possible to suggest that, on the one hand, Brazil occupies an emerging position in the world economic scenario. Such benefits that could come from such prominence end up not being socially distributed when it is noticed that a good portion of society is outside the system.
