Comparing real wage growth across 2001, 2008 and 2020 US recessions

```python
import tradingeconomics as te
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
te.login(); # Signup for a free API key at https://developer.tradingeconomics.com/
```

```python
data_inflation = te.getHistoricalData(country='united states', indicator='inflation rate', initDate='2000-01-01')
data_wages = te.getHistoricalData(country='united states', indicator='wage growth', initDate='2000-01-01')
```

```python
plt.plot(data_inflation, label='Inflation Rate')
plt.plot(data_wages, label='Wage Growth')
plt.title("United States - Inflation and Nominal Wage Growth")
plt.grid(True)
plt.ylabel("%")
plt.xlabel("Date")
plt.legend()
plt.show()
```

Although, in 2021, we register the highest inflation rate after a recession, we also register the biggest rise in real wages for the whole economy.

```python
data_real_wages = data_wages
for i, inflation in enumerate(data_inflation[0]):
    try:
        #wages - inflation = real wages
        data_real_wages[0][i] = data_real_wages[0][i] - inflation
    except:
        print('Index out of bounds.')
```

```python
plt.plot(data_real_wages)
plt.title("United States - Real Wage Growth")
plt.grid(True)
plt.ylabel("%")
plt.xlabel("Date")
plt.show()
```
