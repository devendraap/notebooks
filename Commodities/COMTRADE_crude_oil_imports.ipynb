```python
import tradingeconomics as te
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
te.login(); # Signup for a free API key at https://developer.tradingeconomics.com/
```

```python
data_1 = te.getCmtHistorical(symbol = 'USASAU27091', output_type='df')
```

```python
dates = pd.DatetimeIndex(data_1['date'])
labels = dates.year
values = data_1['value']
plt.style.use('ggplot')
x = labels
crude = values
x_pos = [i for i, _ in enumerate(x)]
plt.bar(x_pos,crude, width=0.8, bottom=None, align='center', data=None,  color='black')
plt.xlabel("year")
plt.ylabel("USD Million")
plt.title("United States Imports from Saudi Arabia of Crude Oil")
plt.ylim(10)
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
plt.xticks(x_pos, x)
plt.tight_layout()
plt.show()
```
