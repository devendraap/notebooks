```python
# Disable warnings in Anaconda
import warnings
warnings.filterwarnings("ignore")
# Matplotlib forms basis for visualization in Python
import matplotlib.pyplot as plt
# We will use the Seaborn library
import seaborn as sns
sns.set()
# Graphics in retina format are more sharp and legible
%config InlineBackend.figure_format = 'retina'
# Increase the default plot size and set the color scheme
import pandas as pd
import tradingeconomics as te
te.login()
```

## Getting Commodities Groups

```python
df_commodities = te.getMarketsData('commodities', 'df')
useful_columns = df_commodities[['Name','Group','DailyPercentualChange','WeeklyPercentualChange', 'MonthlyPercentualChange','YearlyPercentualChange','YTDPercentualChange']]
useful_columns=useful_columns.rename(columns={'Name':'Commodity'} )
```

## Group - Energy

```python
df = useful_columns.query(f'Group =="Energy"')
df = (
df.pivot_table(
        index="Commodity", values=["DailyPercentualChange","WeeklyPercentualChange","MonthlyPercentualChange","YearlyPercentualChange",'YTDPercentualChange'], aggfunc=sum
    )
    .fillna(0)
    .applymap(float)
)
plt.rcParams["figure.figsize"] = 10, 8
plt.rcParams["image.cmap"] = "viridis"
sns.heatmap(df, annot=True, fmt=".1f", linewidths=0.5);
    
```
